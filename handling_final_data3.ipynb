# 아파트 가치평가 지표들 생성

## 기존 데이터 프레임에서 0값들 제거

```python
# 저장한 파일 불러오기
final_df = pd.read_pickle('/content/drive/MyDrive/house_price/final_df.pkl')
final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop(final_df[(final_df.deal_price == 0)|(final_df.full_rent_price == 0)|(final_df.year_rent_price == 0)].index, inplace=True)
final_df.reset_index(drop=True,inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

## 평가지표 컬럼들 추가

```python
# 전세가율, 연세가율을 구함함
final_df['deal_to_full_rent_rate'] = final_df['deal_price']/final_df['full_rent_price']
final_df['deal_to_year_rent_rate'] = final_df['deal_price']/final_df['year_rent_price']
final_df
```

```python
final_df.info()
```

```python
# 소수점 자리 수정정
final_df['deal_to_full_rent_rate'] = final_df['deal_to_full_rent_rate'].astype(float).round(2)
final_df['deal_to_year_rent_rate'] = final_df['deal_to_year_rent_rate'].astype(float).round(2) # 반올림이 되지 않아서 따로 실행행
```

```python
# 타입변경
final_df = final_df.astype({'deal_to_full_rent_rate':'float32', 'deal_to_year_rent_rate':'float32'})
final_df
```

```python
# null 값들과 inf 값들 있는지 확인인
import numpy as np
print(final_df.isnull().sum())

final_df.replace([np.inf, -np.inf], np.nan, inplace=True)
var = final_df.isnull().sum()
print(var.to_string())
```
